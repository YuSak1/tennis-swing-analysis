```python
# Replace outliers
def replace_outliers(df):
  for column in df:
    mean = df[column].mean()
    std = df[column].std()
    lower_bound = mean - 3*std
    upper_bound = mean + 3*std
    df[column] = df[column].apply(lambda x: mean if x > upper_bound or x < lower_bound else x)
  return df

# Normalization
def normalization(df):

  df_x = df.loc[:, ['left_shoulder_x', 'right_shoulder_x', 'left_elbow_x', 'right_elbow_x', 'left_wrist_x',
                    'right_wrist_x', 'left_hip_x', 'right_hip_x', 'left_knee_x', 'right_knee_x', 'left_ankle_x', 'right_ankle_x']]
  df_y = df.loc[:, ['left_shoulder_y', 'right_shoulder_y', 'left_elbow_y', 'right_elbow_y', 'left_wrist_y',
                    'right_wrist_y', 'left_hip_y', 'right_hip_y', 'left_knee_y', 'right_knee_y', 'left_ankle_y', 'right_ankle_y']]

  df_x = df_x.apply(lambda iterator: ((df.max().max() - iterator)/(df.max().max() - df.min().min())).round(5))
  df_y = df_y.apply(lambda iterator: (1- (df.max().max() - iterator)/(df.max().max() - df.min().min())).round(5))
  df_out = pd.concat([df_x, df_y], axis=1)

  # Replace outliers (better not to use this)
  # df_out = replace_outliers(df_out)

  return df_out
```

```python
def create_features(df):

  df_out = pd.DataFrame()

  # Good features (above 70% acc)
  df_out['left_wrist_up'] = df['left_wrist_x'] - df['left_shoulder_x'] #78.0%
  df_out['right_wrist_up'] = df['right_wrist_x'] - df['right_shoulder_x'] #73.1%
  df_out['right_elbow_dist'] =  df['right_elbow_y'] - (df['right_shoulder_y'] + df['left_shoulder_y'])/2 #71.1%
  df_out['right_arm_y'] =  df['right_shoulder_y'] - df['right_wrist_y'] #70.0

  # Medium features (40-70% acc)
  df_out['left_arm_y'] =  df['left_shoulder_y'] - df['left_wrist_y'] #67.8%
  df_out['wrist_y'] = df['right_wrist_y'] - df['left_wrist_y'] #65.8%
  df_out['shoulder_y'] = df['right_shoulder_y'] - df['left_shoulder_y'] #60.8%
  # df_out['right_elbow_open'] =  df['right_elbow_y'] - (df['right_shoulder_y'] + df['right_hip_y'])/2 #58.4%
  df_out['left_elbow_dist'] = (df['right_shoulder_y'] + df['left_shoulder_y'])/2 - df['left_elbow_y'] #58.1%
  # df_out['left_elbow_open'] = (df['left_shoulder_y'] + df['left_hip_y'])/2 - df['left_elbow_y'] #55.3%
  df_out['wrist_ankle_y'] = df['right_wrist_y'] - df['left_ankle_y'] #54.7%
  # df_out['left_elbow_up'] = df['left_elbow_x'] - df['left_shoulder_x'] #50.8%
  # df_out['right_elbow_up'] = df['right_elbow_x'] - df['right_shoulder_x'] #48.2%
  # df_out['left_ankle_dist'] = (df['right_shoulder_y'] + df['left_shoulder_y'])/2 - df['left_ankle_y'] #42.7%
  df_out['ankle_open'] = df['right_ankle_y'] - df['left_ankle_y'] #42.3%

  # Bad features (below 40% acc)
  # df_out['center_y'] = ((df['right_ankle_y']+df['left_ankle_y'])/2) - ((df['right_hip_y']-df['left_hip_y'])/2) #39.3%
  # df_out['right_ankle_dist'] =  df['right_ankle_y'] - (df['right_shoulder_y'] + df['left_shoulder_y'])/2 #39.1%
  # df_out['hip_y'] = df['right_hip_y'] - df['left_hip_y'] #29.7%
  # df_out['right_hip_hight'] = df['right_hip_x'] - df['right_ankle_x'] #28.5%
  # df_out['hip_hight'] = ((df['right_hip_x'] - df['right_ankle_x']) + (df['left_hip_x'] - df['left_ankle_x']))/2 #25.9

  return df_out
```

```python
def expand_data(df_input, num_frames: int, num_samples: int, print_s=True):

  input_size = df_input.shape[0]

  if int(input_size / num_frames) + 1 > num_samples:
    print("num_samples must be larger than ", int(input_size / num_frames), " !")

  stride = ((input_size - num_frames) // num_samples)
  if print_s==True:
    print("stride: ", stride)

  df_expand = pd.DataFrame()
  for i in range(0, input_size - num_frames + 1, stride):
    df = df_input.iloc[i:i+num_frames]
    df_expand = pd.concat([df_expand, df])

  df_expand = df_expand.reset_index(drop=True)

  return df_expand
```

```python
# Set the number of frames for each sample, and the number of samples for training and test data
num_frames = 90
# num_samples_train = 1200
num_samples_train = 1500
num_samples_test = 400

# Expand the data set of each players
df_federer_train_expand = expand_data(df_federer_train, num_frames, num_samples_train)
df_nadal_train_expand = expand_data(df_nadal_train, num_frames, num_samples_train)
df_djokovic_train_expand = expand_data(df_djokovic_train, num_frames, num_samples_train)
df_murray_train_expand = expand_data(df_murray_train, num_frames, num_samples_train)
print('=====================')
df_federer_test_expand = expand_data(df_federer_test, num_frames, num_samples_test)
df_nadal_test_expand = expand_data(df_nadal_test, num_frames, num_samples_test)
df_djokovic_test_expand = expand_data(df_djokovic_test, num_frames, num_samples_test)
df_murray_test_expand = expand_data(df_murray_test, num_frames, num_samples_test)
```

```python
# v3.0 (final version)
num_classes = 4

def sub_model(input_shape):
  input_layer = Input(input_shape)

  conv = Conv1D(filters=64, kernel_size=9, padding="same")(input_layer)
  conv = BatchNormalization()(conv)
  conv = ReLU()(conv)
  conv = Dropout(0.25)(conv)

  conv = Conv1D(filters=64, kernel_size=9, padding="same")(conv)
  conv = BatchNormalization()(conv)
  conv = ReLU()(conv)
  conv = Dropout(0.25)(conv)

  conv = Conv1D(filters=64, kernel_size=9, padding="same")(conv)
  conv = BatchNormalization()(conv)
  conv = ReLU()(conv)
  conv = Dropout(0.25)(conv)

  gap = GlobalAveragePooling1D()(conv)
  dense = Dense(256)(gap)
  model = Model(inputs=input_layer, outputs=dense)

  return model


def make_model(input_shape):

  model1 = sub_model(input_shape)
  model2 = sub_model(input_shape)
  model3 = sub_model(input_shape)
  model4 = sub_model(input_shape)
  model5 = sub_model(input_shape)
  model6 = sub_model(input_shape)
  model7 = sub_model(input_shape)
  model8 = sub_model(input_shape)
  model9 = sub_model(input_shape)
  model10 = sub_model(input_shape)

  # concatenate
  combined = concatenate([model1.output, model2.output, model3.output, model4.output, model5.output, model6.output, model7.output, model8.output, model9.output, model10.output])
  dense = Dense(256)(combined)
  dense = Dense(128)(dense)
  dense = Dense(64)(dense)
  output_layer = Dense(num_classes, activation="softmax")(dense)

  return Model(inputs=[model1.input, model2.input, model3.input, model4.input, model5.input, model6.input, model7.input, model8.input, model9.input, model10.input], outputs=output_layer)

model = make_model(input_shape=(x_train.shape[1], 1))
# keras.utils.plot_model(model, show_shapes=True)
```

# Plot confusion matrix

```python
from sklearn.metrics import confusion_matrix
import itertools

def plot_confusion_matrix(cm, classes,
                          normalize=False,
                          title='Confusion matrix',
                          cmap=plt.cm.Blues):
    """
    This function prints and plots the confusion matrix.
    Normalization can be applied by setting `normalize=True`.
    """
    if normalize:
        cm = cm.astype('float') / cm.sum(axis=1)[:, np.newaxis]
        print("Normalized confusion matrix")
    else:
        print('Confusion matrix, without normalization')

    print(cm)

    plt.imshow(cm, interpolation='nearest', cmap=cmap)
    plt.title(title)
    plt.colorbar()
    tick_marks = np.arange(len(classes))
    plt.xticks(tick_marks, classes, rotation=45)
    plt.yticks(tick_marks, classes)

    fmt = '.2f' if normalize else 'd'
    thresh = cm.max() / 2.
    for i, j in itertools.product(range(cm.shape[0]), range(cm.shape[1])):
        plt.text(j, i, format(cm[i, j], fmt),
                 horizontalalignment="center",
                 color="white" if cm[i, j] > thresh else "black")

    plt.tight_layout()
    plt.ylabel('True label')
    plt.xlabel('Predicted label')


pred_y = model.predict([x_test[:,:,0], x_test[:,:,1], x_test[:,:,2], x_test[:,:,3], x_test[:,:,4], x_test[:,:,5], x_test[:,:,6], x_test[:,:,7], x_test[:,:,8], x_test[:,:,9]])
# pred_y = model.predict(x_test)
pred_y_classes = np.argmax(pred_y,axis = 1)
true_y= np.argmax(y_test, axis = 1)
confusion_mtx = confusion_matrix(true_y, pred_y_classes)
plot_confusion_matrix(confusion_mtx, classes = ['F', 'N', 'D', 'M'])
```
